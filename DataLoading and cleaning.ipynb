```python
table1 =  pd.merge(content, reactions, on=['Content ID'], how = 'inner')
table1.drop(['Unnamed: 0_x','Unnamed: 0_y'], axis = 1, inplace = True)
table1
```

```python
table1.columns
```

```python
table1.rename(columns = {'User ID_x':'User_ID_c', 'Type_x': 'Type_c', 'User ID_y': 'User_ID_r', 'Type_y': 'Type'},inplace= True )
table1
```

```python
finaltable = pd.merge(table1, reactionTypes, on= ['Type'], how='inner')
finaltable.drop(['Unnamed: 0'], axis = 1, inplace = True)
finaltable
```
